```python
raw_data['math_cut']=pd.cut(raw_data['math score'].apply(int),bins=[0,60,80,90,100],right=True,labels=['<60','60-80','80-90','>90'])
raw_data[['math_cut','math score']]
```

```python
raw_data['reading_cut']=pd.cut(raw_data['reading score'].apply(int),bins=[0,60,80,90,100],right=True,labels=['<60','60-80','80-90','>90'])
raw_data[['reading_cut','reading score']]
```

```python
raw_data['writing_cut']=pd.cut(raw_data['writing score'].apply(int),bins=[0,60,80,90,100],right=True,labels=['<60','60-80','80-90','>90'])
raw_data[['writing_cut','writing score']]
```

```python
p1=raw_data [(raw_data['math score'] >90) & (raw_data['reading score'] >90) & (raw_data['writing score'] >90)]
p2= p1.groupby(['race/ethnicity','parental level of education']).count().reset_index()
```
